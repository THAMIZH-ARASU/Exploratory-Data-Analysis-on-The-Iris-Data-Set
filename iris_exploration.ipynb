```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
# Display the first few rows of the dataset
print("First five rows of the dataset:")
print(df.head())
```

```python
# Basic statistics and summary
print("\nBasic statistics and summary:")
print(df.describe())
```

```python
# Checking for missing values
print("\nChecking for missing values:")
print(df.isnull().sum())
```

```python
# Distribution of the species
print("\nDistribution of species:")
print(df['species'].value_counts())
```

# Data Visualisation

```python

# Histograms for each feature
print("\nPlotting histograms for each feature:")
df.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.suptitle("Histograms of Features")
plt.show()
```

```python
# Pairplot to visualize pairwise relationships
print("\nPlotting pairplot for pairwise relationships:")
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
# Boxplot for each feature
print("\nPlotting boxplot for each feature:")
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, width=0.5, fliersize=5)
plt.title("Boxplot of Features")
plt.show()
```

```python
# Violin plot for each feature
print("\nPlotting violin plot for each feature:")
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='sepal length (cm)', data=df)
plt.title("Violin Plot of Sepal Length")
plt.show()
```

```python
# Heatmap for correlation analysis
print("\nPlotting heatmap for correlation analysis:")
correlation_matrix = df.drop(columns = ['species']).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Pairwise correlation of features
print("\nPairwise correlation of features:")
print(correlation_matrix)
```

```python
# Scatter plot with regression line for Sepal Length vs Sepal Width
print("\nScatter plot with regression line for Sepal Length vs Sepal Width:")
sns.lmplot(x='sepal length (cm)', y='sepal width (cm)', data=df, hue='species', markers=["o", "s", "D"])
plt.title("Sepal Length vs Sepal Width with Regression Line")
plt.show()
```

# Overall Conclusion:

## The EDA of the Iris dataset reveals that:
1. Iris setosa is easily distinguishable from the other species based on its feature values, particularly petal length and petal width.
2. Iris versicolor nd Iris virginica have overlapping feature distributions, making them harder to distinguish, although petal measurements still provide significant discriminatory power.
3. Petal length and petal width are the most useful features for differentiating between the species, while sepal width is the least useful.
4. There are no missing values, and the dataset is balanced across the three species.
